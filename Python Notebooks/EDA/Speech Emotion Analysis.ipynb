## Let's observe the wave patterns for Anger, Sad and Happy emotions in the given in one of each audio files

```python
import pyaudio  
import wave  
import librosa
import matplotlib.pyplot as plt
```

```python
# Anger
y1,sr1 = librosa.load('../IEMOCAP_full_release/Session1/sentences/wav/Ses01F_impro01/Ses01F_impro01_F012.wav')
#sad
y2,sr2 = librosa.load('../IEMOCAP_full_release/Session1/sentences/wav/Ses01F_impro06/Ses01F_impro06_M003.wav')
# happy
y3,sr3 = librosa.load('../IEMOCAP_full_release/Session1/sentences/wav/Ses01F_impro03/Ses01F_impro03_F008.wav')
```

```python
plt.figure(figsize=(15,2))
librosa.display.waveshow(y1, sr=sr1, alpha=0.25, color='r')
```

```python
plt.figure(figsize=(15,2))
librosa.display.waveshow(y2, sr=sr2, alpha=0.25, color='g')
```

```python
plt.figure(figsize=(15,2))
librosa.display.waveshow(y3, sr=sr3, alpha=0.25, color='b')
```

## Root Mean Square (RMS)

### RMS is a quantitative measure and doesn't directly correspond to how humans perceive loudness. However, it does provide a useful metric for comparing the relative amplitude of different audio signals.

### In emotion recognition applications, the "loudness" or intensity captured by RMS is just one of many features considered. Emotional expression in speech involves a complex interplay of various acoustic features, including pitch, prosody, duration, and more

```python
# Calculate the RMS
rmse1 = librosa.feature.rms(y=y1)
rmse2 = librosa.feature.rms(y=y2)
rmse3 = librosa.feature.rms(y=y3)

print(rmse1)
print(rmse2)
print(rmse3)
```

```python
# Display the RMS over time for Anger
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y1, sr=sr1, alpha=0.5)
plt.plot(librosa.times_like(rmse1), rmse1[0], label='RMS')
plt.title('Waveform and RMS - Anger/Frustation')
plt.legend()
plt.show()
```

```python
# Display the RMS over time for Sad
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y2, sr=sr2, alpha=0.5, color='r')
plt.plot(librosa.times_like(rmse2), rmse2[0], label='RMS')
plt.title('Waveform and RMS - Sad')
plt.legend()
plt.show()
```

```python
# Display the RMS over time for Happy
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y3, sr=sr3, alpha=0.5, color='g')
plt.plot(librosa.times_like(rmse3), rmse3[0], label='RMS')
plt.title('Waveform and RMS - Happy')
plt.legend()
plt.show()
```
