## Sine Wave

```python
# simple sine wave
A = 1                      # Amplitude
f0 = 1                     # frequency
Fs = 1600                  # Sampling frequency
t = np.arange(0, 1, 1/Fs) 

X = A * np.sin(2 * np.pi * f0 * t)

plt.plot(t, X)
plt.title("Sine Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

## What is complex wave and how it look likes

```python
# create two independent waves then adding them resulting wave is complex wave.

# wave 1
A = 1/3
f0 = 20
Fs = 1600
t = np.arange(0, 1, 1/Fs)
X1 = A * np.sin(2  * np.pi * f0 * t)

# wave 2
A = 0.5
f0 = 3
Fs = 1600
t = np.arange(0, 1, 1/Fs)
X2 = A * np.cos(2  * np.pi * f0 * t)

# wave 3 = wave 1 + wave 2
X3 = X1 + X2

plt.figure(figsize=(10,5))
plt.title("Complex wave")
plt.plot(X1)
plt.plot(X2)
plt.plot(X3)
plt.legend(['x1','x2','x3'])
plt.show()
```

## What is Audio Speech and how it looks like

```python
data, sampling_rate = librosa.load('../IEMOCAP_full_release/Session1/sentences/wav/Ses01F_impro01/Ses01F_impro01_F005.wav')
# sampling frequency nothing but how may samples present for second which is similar to sampling rate
print(f"Sampling frequency : {sampling_rate} and Wave : {data}")
print()
```

## Time vs Frequency domain


```python
samplerate=22050
times = np.arange(len(data))/float(sampling_rate) # Calculate the duration of the signal

print('Minimum time (sec):', min(times))
print('Maximum time (sec):', max(times))
print('Minimum data point value:', min(data))
print('Maximum data point value:', max(data))

duration = len(data)/float(samplerate)

print('Duration of the audio in seconds: {:.2f}'.format(duration))

print('Number of samples:',len(data))
```

### Time Domain

```python
plt.plot(data)
plt.title("Time Domain Representation (Speech Utterance Waveform)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

### Frequency Domain

```python
from scipy.fft import fft, rfft

# Apply the FFT on the signal
fourier = fft(data)

# Plot the result (the spectrum |Xk|)
plt.plot(np.abs(fourier))
plt.title("Frequency domain representation-Output of Fast Forier Transform FFT")
plt.xlabel("Number of Samples")
plt.ylabel("Amplitude of Sinusoids")

plt.show()
```

```python
# Calculate N/2 to normalize the FFT output
N = len(data)
normalize = N/2

# Plot the normalized FFT (|Xk|)/(N/2)
plt.plot(np.abs(fourier)/normalize)
plt.ylabel('Amplitude')
plt.xlabel('Samples')
plt.title('Normalized FFT Spectrum')
plt.show()
```

### Spectrum 
### The spectrum with the actual amplitudes and frequency axis. 


```python
from scipy.fft import fftfreq, rfftfreq
# Get the frequency components of the spectrum
frequency_axis = fftfreq(N, d=1.0/sampling_rate)
norm_amplitude = np.abs(fourier)/normalize
# Plot the results
plt.plot(frequency_axis, norm_amplitude)
plt.xlabel('Frequency[Hz]')
plt.ylabel('Amplitude')
plt.title('Spectrum')
plt.show()
```

### Spectrogram

```python
X = librosa.stft(data)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(20, 5))
librosa.display.specshow(Xdb, sr=sampling_rate, x_axis='time', y_axis='hz')
plt.title('Spectrogram')
plt.colorbar()
```

```python
#convert the frequency axis to a logarithmic one.
librosa.display.specshow(Xdb, sr=sampling_rate, x_axis='time', y_axis='log')
plt.colorbar()
```

### Mel Spectrogram

```python
#Mel spectrogram
# Extract mel spectrogram features 
S = librosa.feature.melspectrogram(y=data, sr=sampling_rate)
S_dB = librosa.power_to_db(S, ref=np.max)

# Display mel spectrogram
fig, ax = plt.subplots()
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sampling_rate,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='Mel-frequency spectrogram')

plt.show()
```

### Creating a simple audio analysis on one of the sample file

```python

threshold = 0.005
print(len('../IEMOCAP_full_release/Session1/sentences/wav/Ses01F_impro01/Ses01F_impro01_F005.wav'))
# Load the audio file
y, sr = librosa.load('../IEMOCAP_full_release/Session1/sentences/wav/Ses01F_impro01/Ses01F_impro01_F005.wav')

# Compute the short-term power spectrogram
S = librosa.feature.melspectrogram(y=y, sr=sr)

# Convert to decibels (dB)
S_dB = librosa.power_to_db(S, ref=np.max)

# Calculate the mean energy in the signal
mean_energy = np.mean(S_dB)
# Visualize the audio waveform and spectrogram
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
librosa.display.waveshow(y, sr=sr)
plt.title('Audio Waveform')

plt.subplot(3, 1, 2)
librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=sr, cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')

plt.subplot(3, 1, 3)
librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=sr, cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.axhline(y=threshold, color='r', linestyle='--', label='Threshold')
plt.title('Mel Spectrogram with Threshold')

plt.tight_layout()
plt.show()
```

## Check if there is any noise present in audio

```python
if(mean_energy > threshold):
     print("Noise is present in the audio file.")
else:
    print("The audio file is relatively clean.")
```
